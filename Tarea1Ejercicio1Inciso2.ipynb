Importación de paquetes y funciones útiles

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

Definición de variables y cración de la grilla

```python
a = 1
b = 1
n = 20 # nodos de la grilla
max = 2 # máximo y mínimo del gráfico
X = np.linspace(-max,max,n)
Z = np.linspace(-max,max,n)
x,z = np.meshgrid(X,Z)
```

Defininicion de las componentes del campo eléctrico

```python
def Ex(x,z):
    return -np.log(np.abs((np.sqrt((-b/2)**2+(x-a/2)**2+z**2)-b/2))/np.sqrt((x-a/2)**2+z**2))+np.log(np.abs((np.sqrt((b/2)**2+(x-a/2)**2+z**2)+b/2))/np.sqrt((x-a/2)**2+z**2))+np.log(np.abs((np.sqrt((-b/2)**2+(x+a/2)**2+z**2)-b/2))/np.sqrt((x+a/2)**2+z**2))-np.log(np.abs((np.sqrt((b/2)**2+(x+a/2)**2+z**2)+b/2))/np.sqrt((x+a/2)**2+z**2))
def Ez(x,z):
    return np.arctan(((-b/2)*(x-a/2))/(z*np.sqrt((x-a/2)**2+(-b/2)**2+z**2)))-np.arctan(((+b/2)*(x-a/2))/(z*np.sqrt((x-a/2)**2+(+b/2)**2+z**2)))-np.arctan(((-b/2)*(x+a/2))/(z*np.sqrt((x+a/2)**2+(-b/2)**2+z**2)))+np.arctan(((+b/2)*(x+a/2))/(z*np.sqrt((x+a/2)**2+(+b/2)**2+z**2)))
```

Creación del gráfico

```python
plt.figure(figsize=(10,10))
plt.quiver(x,z,Ex(x,z)/E(x,z),Ez(x,z)/E(x,z),E(x,z), pivot='middle', cmap='winter_r')
plt.xlim(-max,max)
plt.ylim(-max,max)
plt.xlabel('$x$')
plt.ylabel('$z$')
plt.colorbar()
plt.title('Representación de $\\vec{E}(x,y,z)$ en el plano $xz$')
```
